```python
# import libraries 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Supress warning for clean notebook
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("C:\\Users\\GBENGA\\Documents\\Melbourne_housing_FULL.csv")
df.head(10)
```

```python
print(f"{df.nunique()} and {df.shape}")
```

```python
col_to_use =['Suburb','Rooms',"Type",'Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2',
            'Bathroom','Car','Landsize','BuildingArea','Price']
dataset = df[col_to_use]
dataset
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero =['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
# filling the dataset with the mean of each column
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price',axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

```python
# this dataset was not the best to explain fitting over fitting and balanced fitting
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg =Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```
